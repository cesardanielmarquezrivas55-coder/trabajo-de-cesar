```python
!pip install gtts

from gtts import gTTS
from IPython.display import Audio

# Escribimos el texto a convertir
texto = "los primeros 100 pokemon son basura pero gana nostalgia"

# hacemos el audio
tts = gTTS(text=texto, lang='es')

# obtenemos y guardamos el audio
tts.save("audio.mp3")

print("Audio generado correctamente.")

# esta parte reproducce el audio
Audio("audio.mp3")
```
