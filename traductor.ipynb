#Traductor 😊

##<b>definimos las palabras del diccionario</b>

```python
dic = {"hola": "hello", "mundo": "word","estrella": "star", "mochila": "backpack", "Nube": "cloud", "casa": "house", "hermano": "brother", "hermana": "sister", "mamá": "mom", "raton": "mouse"}
```

##<b>establecemos para que regrese la palabra que esta marcada como igual y regrese la contraria</b>

```python
def translate(word):
    return dic.get(word, "no encontrado")
```

##<b>escojemos las palabras que va a traduccir</b>

```python
word = "estrella"
```

##<b>imprimimos la palabra</b>

```python
print(translate(word))
```
