```python
import speech_recognition as sr

filename = "audio.wav"
output_file = "transcripcion_audio.txt"

r = sr.Recognizer()

with sr.AudioFile(filename) as source:
    print("analizando el archivo...")
    audio_data = r.record(source)

try:
    print("convirtiendo a escrito...")
    text = r.recognize_google(audio_data, language="es-MX")

    with open(output_file, "w") as f:
        f.write(text)

    print("Ya esta puedes revisar los resultados")
    print("Contenido:", text)

except Exception as e:
    print(f"que bot te equivocaste: {e}")
```
